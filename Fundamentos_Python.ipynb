```python
'''Escreva e formate a data em que você nasceu no formato dia/mes/ano.
Não se esqueça de criar 3 variaveis para guardar o dia, mes e ano.'''
from datetime import datetime, date

dataNascimento = input("Informe sua data de nascimento")
convertDtNascimento = datetime.strptime(dataNascimento, "%d/%m/%Y")
diaNascimento = convertDtNascimento.day
mesNascimento = convertDtNascimento.month
anoNascimento = convertDtNascimento.year

print(f"Você nasceu em: {diaNascimento}/{mesNascimento}/{anoNascimento}")
```
